```python
import pandas as pd


# 讀取 CSV 並確保找得到缺失值
df = pd.read_csv("/Users/wanghao/fa_econ_group6/fa-25-econ-5166-group-6/data/raw/data112.csv", na_values=["", " ", "NA", "NaN", "null"])


# 保留指定欄位
df = df[['scale','a1','a2','a3','a4','a5','a6','a7','a8','a9','a10','a11','a12',
         'b1a','b6','b10','b13','b14','b16','c1','d3','e1','e1a_1','e1a_2','e1a_3',
         'f1','f2','f3','f5','f7','f8a','f8b','f8c','f8d','f9']]

print(df.isnull().sum())
```

### 處理變項

```python
# 將 b1a 的空白值補 0
df["b1a"] = df["b1a"].replace(r"^\s*$", 0, regex=True)  # 空白或空字串補 0
df["b1a"] = pd.to_numeric(df["b1a"], errors="coerce").fillna(0).astype(int)

# 將 b10 欄位中 2 和 3 改成 1，1 改成 0 1為可以遠距工作，0為不行
df['b10'] = df['b10'].replace({2: 1, 3: 1, 1: 0})



# 建立新欄位，將f8a乘上12加上f8b 命名為 f8a_month(目前工作單位)，將f8c乘上12加上f8d 命名為 f8b_month（總年資）
df["f8a_month"] = df["f8a"] * 12 + df["f8b"]
df["f8b_month"] = df["f8c"] * 12 + df["f8d"]



# 移除原本欄位
df.drop(columns=["f8a", "f8b", "f8c", "f8d"], inplace=True)
```

### e1a畫長條圖

```python
import pandas as pd
import matplotlib.pyplot as plt


# 合併三個欄位
all_values = pd.concat([df["e1a_1"], df["e1a_2"], df["e1a_3"]])

# 計算相對頻率
relative_freq = all_values.value_counts(normalize=True).sort_index()

# 畫圖
ax = relative_freq.plot(kind='bar')

# ✅ 改成英文標籤與標題
plt.xlabel("Category (1–10)", fontsize=12)
plt.ylabel("Relative Frequency", fontsize=12)
plt.title("Relative Frequency Distribution of e1a_1 ~ e1a_3", fontsize=14)

# ✅ 讓 X 軸標籤直立顯示（避免亂碼 & 重疊）
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 移除原本欄位
df.drop(columns=['e1a_1','e1a_2','e1a_3'], inplace=True)
```

```python
# 將處理好的 df 存成 CSV
df.to_csv("/Users/wanghao/fa_econ_group6/fa-25-econ-5166-group-6/data/processed/processed_data112.csv", index=False)
```
